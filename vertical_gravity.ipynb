```python
import matplotlib.pyplot as plt
import numpy as np

#Prompt for initial height
h = eval(input("What is the initial height in meters? "))

#Prompt for initial velocity
v = eval(input("What is the initial velocity in m/s? "))

#Create Graph
#n = 400 # Number of Points
x = np.linspace(0, 25, 400) # t-axis

#Create Equations
earth = -5*x**2 + v*x + h
mars = -2*(x**2) + v*x + h
moon = -(x**2) + v*x + h
jupiter = -12.5*(x**2) +v*x + h

#Plot Grpahs
plt.xlabel('time (seconds)')
plt.ylabel('height (meters)')
plt.ylim(0, 120)
plt.xlim(0,25)
plt.title('Vertical Motion under Graviity on 3 Planets')

plt.plot(x, earth, 'r-', label='Earth') # plotting t, earth separately 
plt.plot(x, mars, 'g--', label='Mars') # plotting t, mars separately 
plt.plot(x, moon, 'b-.', label='Moon') # plotting t, moon separately 
plt.plot(x,jupiter, 'k-')

plt.legend(["Earth", "Mars", "Moon"], loc ="upper right")


```
